```python
# Import dependencies
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Import our custom ImageStitcher class
from imagestitcher import ImageStitcher
```

```python
# Helper function to display images inline
def show_image(img, title="Image", figsize=(12, 8)):
    """Display an image using matplotlib."""
    plt.figure(figsize=figsize)
    if len(img.shape) == 3:
        plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```
